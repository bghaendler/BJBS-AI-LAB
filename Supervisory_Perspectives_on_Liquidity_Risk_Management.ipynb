# CHAPTER 10 Supervisory Perspectives on Liquidity Risk Management

## INTRODUCTION

- Financial institutions must understand their liquidity needs at an enterprise-wide level and be prepared for unexpected erosion of market liquidity.
- The global liquidity crisis in 2007-08 highlighted challenges faced by banking organizations and renewed attention to liquidity risk management.
- Banking supervisors have highlighted failures of firms to take basic principles of liquidity risk management into account when liquidity was plentiful.
- Regulators have renewed emphasis on liquidity risk management in regulations such as the Enhanced Prudential Standards (EPS) and standards issued by the Basel Committee on Banking Supervision (BCBS).
- Regulatory and supervisory requirements complement the quantitative measures implemented under the Basel III framework.
- The chapter will review the evolution and range of regulatory and supervisory requirements relative to liquidity risk management, including firms' governance and internal controls frameworks.

## RATING LIQUIDITY RISK MANAGEMENT WITH U.S. BANKING REGULATORS’ RATING SYSTEM

- The CAMEL rating system was developed in 1979 to provide a uniform rating system for banks in the US.
- The system covers financial, governance, and internal controls measures, with a focus on liquidity risk management practices.
- Liquidity position, funding needs, and adequacy of funds management practices are important considerations in evaluating a financial institution's liquidity.
- Banks should be able to manage unplanned changes in funding sources and react to market conditions affecting asset liquidation.
- The board of directors of a bank should provide clear guidance on acceptable risk exposure levels and ensure the establishment of appropriate policies and procedures.
- Regulatory requirements and supervisory expectations have evolved over time in response to financial market events, aiming to address certain practices in supervised banking organizations.
- The assessment of management plays a crucial role in the CAMEL rating system, including the identification, measurement, monitoring, and control of liquidity position.
- Regulatory releases from banking supervisors become increasingly prescriptive over time, shaping the framework of requirements and expectations.

## FOUNDATIONS ESTABLISHED IN BCBS’ “SOUND PRACTICES FOR MANAGING LIQUIDITY IN BANKING ORGANIZATIONS”

1. Introduction:
    - BCBS (Bank for International Settlements' Basel Committee on Banking Supervision) issued a paper titled "Sound Practices for Managing Liquidity in Banking Organizations" in response to the new millennium.
    - Banking supervisors worldwide aimed to strengthen liquidity risk management practices in supervised firms.
2. Emerging Trends:
    - Firms increasingly relied on wholesale funding markets and utilized new financial and technological innovations for funding and managing liquidity exposures.
    - The collapse of Long Term Capital Management and the Russian debt moratorium triggered a liquidity crisis.
    - The "flight to quality" response caused unexpected consequences in prevailing funding models, resulting in a spiral effect for assets as market participants struggled to meet margin calls.
3. BCBS Paper:
    - BCBS issued the paper to replace an existing 1992 paper on liquidity risk management.
    - The new paper focused on fourteen principles for effective liquidity risk management within supervised firms.
4. Subsequent Evolution:
    - The following sections will discuss the established principles in the BCBS paper and their evolution through subsequent guidance and regulations issued by banking supervisors.

## STRATEGY SETTING AND THE OVERSIGHT ROLE OF DIRECTORS AND SENIOR MANAGEMENT

1. Strategy Setting and Oversight
    - Importance of effectively setting strategies for liquidity risk management
    - Supervisory focus on quantitative and qualitative targets
    - Need to withstand stressed market conditions in short- and long-term
2. Role of Board of Directors
    - Approving firm's strategy for managing liquidity risk
    - Approving policies and procedures for managing liquidity risk
    - Monitoring exposures relative to the defined strategy
    - Reviewing timely and detailed information on liquidity risk
    - More frequent reviews for firms with significant funding concentrations or composition changes
    - Regulation YY requiring semiannual reporting by CRO/senior management to the board
3. Role of Senior Management
    - Policy decision making and review of liquidity decisions
    - Authority delegated by the board
    - Oversight through committee structure (ALCO), central treasury, or risk management
    - Maintaining checks and balances for oversight activities
    - Setting limits for cash flow mismatches, reduced marketability, drawdowns, and liquid assets
4. Liquidity Risk Tolerance
    - Bank's need for an appropriate liquidity risk tolerance aligned with business strategy
    - Annual approval of liquidity risk tolerance strategy
    - Incorporating liquidity costs, benefits, and risks in pricing, performance measurement, and new product approval processes
5. Governance Expectations for U.S. Firms
    - Reinforcement of BCBS standards by U.S. banking regulators
    - Clear communication of liquidity risk tolerance throughout the institution
    - Board or delegated committee oversight of liquidity management strategies, policies, and procedures
    - Annual reviews of liquidity management frameworks
- **Measuring and Monitoring Exposures**
    
    BCBS 2000:
    
    - Strategy and board of directors' objectives driving principles
    - Monitoring responsibilities for firm's strategy
    - Measurement and monitoring of exposures and net funding requirements
    - Adequacy of systems for controlling and reporting liquidity risks
    - Stress testing capabilities and validation of assumptions
    - Management information systems: flexibility, timely information, currency calculations
    - Monitoring of shorter-term and longer-term horizons
    - Evaluation of liquidity exposures and stress scenarios
    - Conservative estimations of inflows, outflows, and asset discounts
    
    BCBS Principles' Relevance:
    
    - Liquidity Coverage Ratio (LCR) and stress testing in Regulation YY
    - Internal, external, and combined scenarios
    - Historical experience, market conventions, and management judgment
    
    Assumptions:
    
    - Frequent review of behavioral assumptions
    - Increasing granularity of assumptions
    - Relying on a stock of liquid assets to offset uncertainty
    - Cash flow estimation processes and collateral management practices
    
    BCBS 2008 and Regulation YY:
    
    - Addressing complexity of legal entity structures
    - Monitoring and controlling liquidity exposures and funding requirements
    - Consideration of legal, regulatory, and operational limitations
    - Transferability of High-Quality Liquid Assets (HQLA) buffers
    - Active management of collateral positions
    - Segmentation of encumbered and unencumbered assets
    - Monetization of collateral in a timely manner
    - Restrictions on cross-border asset flows in 2009-10
- **Market Access and Contingency Funding**
    
    Liquidity Risk Management:
    
    - Ability to meet financial obligations under normal and adverse market conditions
    
    Critical Elements:
    
    - Managing access to diverse funding sources
    - Maintaining contingency funding plans
    
    Principles from BCBS 2000:
    
    - Managing diversification of funding sources and outflows
    - Capacity to monetize assets
    - Establishing procedures for addressing cash flow needs in crisis situations
    
    Concentration Matters:
    
    - Monitoring concentrations by counterparty and instrument type
    - Considering the nature and potential behavior of funds provider and geographic markets
    - Alluding to regional stress conditions in the late 1990s
    
    Refinements in BCBS 2008 and U.S. Regulators' 2010 Guidance:
    
    - Refining the concept of contingency funding plans
    - Delineating policies for reacting to stress scenarios
    - Roles and responsibilities of management in addressing crises
    - Communications plans and transparency of plan invocation triggers
    - Escalation procedures
    - Regular testing of the plan's effectiveness
    
    Expectations in U.S. Regulators' Guidance:
    
    - Early warning indicators to identify increased liquidity risks or vulnerabilities
    - Recommended risk measures ranging from concentrations to funding costs and capital levels

## FOREIGN CURRENCY LIQUIDITY MANAGEMENT

1. Currency Problems in 1990s:
    - Currency problems in various markets
    - Liquidity issues faced by global banking organizations
2. Extension of Liquidity Risk Management Standards:
    - Regulators extended standards to address funding requirements in major currencies
    - Applies to banks engaging in those currencies
    - Covers exposure reporting, cash flow estimations, stress testing, and contingency funding planning
3. Basis Risk and BCBS Guidance (2008):
    - BCBS guidance extended requirements
    - Incorporate monitoring and management of liquidity risk exposures and funding needs
    - Across legal entities, business lines, and currencies
    - Consider potential limitations to liquidity transferability across jurisdictions
4. Federal Reserve's Regulation YY:
    - Reiterates the liquidity risk management requirement
    - Emphasizes firms' data management capabilities for effective risk reporting to senior management
5. Currency Risks and Intraday Liquidity Management:
    - Regulators emphasize currency risks in intra-day liquidity risk management
    - Firms' capacity to manage outflows within jurisdictions on a global scale

## CORE INTERNAL CONTROLS FOR LIQUIDITY RISK
MANAGEMENT

1. Core internal controls for liquidity risk management:
    - Establishment of a strong control environment
    - Policies and procedures
    - Effective information systems
    - Continual independent review of adherence to policies and procedures
2. Evolution of internal control requirements:
    - BCBS 2000 emphasized internal controls
    - BCBS 2008 and U.S. regulators' 2010 guidance covered internal controls
    - Regulation YY introduced more prescriptive internal control requirements for independent oversight of liquidity risk management controls
3. Challenges in implementing independent liquidity risk management functions:
    - Lack of clearly defined roles and responsibilities
    - Opaque lines between front office activities and independent control functions
4. Increasing expectations for independent oversight:
    - Expertise of staff within independent oversight functions
    - Complexity of products, modeling methodologies, and assumptions requiring specialized expertise
5. Role of internal audit:
    - Previously relied on for independent oversight responsibilities
    - Enhanced independent risk functions in most firms
    - Internal audit now assesses the effectiveness of the first and second lines of defense, adherence to policies and procedures, banking regulations, and emerging control needs.

## THE DISCIPLINE OF PUBLIC DISCLOSURE

1. Crisis events in the late 1990s prompted global banking supervisors to enhance firms' disclosure of information.
    - Objective: Improve transparency, reduce uncertainty, and strengthen market discipline.
    - BCBS 2000: Introduced measures to enhance disclosure.
2. Limited attention and marginal disclosure improvements prior to the financial crisis.
    - Regulators and regulated firms showed minimal interest in implementing the measures.
    - BCBS 2008: Principle reiterated.
3. Release of regulatory metrics: LCR and Net Stable Funding Ratio (NSFR).
    - Purpose: Assure investors, counterparties, and analysts of firms' ability to meet supervisory expectations.
    - Firms have been attentive in releasing reports concerning these metrics.
    - BCBS: Released disclosure guidelines for LCR and NSFR.
    - U.S. regulators: Implemented LCR-related disclosure requirements.

## MONITORING ADHERENCE TO BCBS STANDARDS

1. BCBS 2000: Introduction of consistent evaluation process and standards for adherence to principles across jurisdictions.
2. Focus on Enactment: Global banking supervisors aimed to ensure that the standards were implemented by banking regulators.
3. Shift in Emphasis: BCBS 2008 brought a change in supervisors' focus towards a firm's resilience to liquidity stress due to its systemic importance.
4. US Regulatory Response: US regulators aligned with BCBS standards through more stringent requirements for systemically important banking organizations.
5. Regulation YY: Imposed stricter requirements on systemically important banking organizations in the US.
6. LCR Final Rule: Imposed liquidity coverage ratio requirements for firms with over $250 billion in assets.
7. Horizontal Reviews: Federal Reserve utilized comprehensive liquidity assessment and review (CLAR) to evaluate liquidity risk management practices and funding profiles of supervised firms.
8. Asset Thresholds: CLAR applied to firms with over $50 billion in assets.
9. Assessment of Adherence: The purpose of horizontal reviews was to assess supervised firms' compliance with regulatory requirements.

## CONCLUSION

1. Supervised firms have questioned prescriptive measures by banking supervisors.
2. Regulators have remained consistent with CAMEL ratings framework and BCBS 2000 for liquidity risk management practices and controls.
3. Quantitative requirements (LCR, NSFR) and the U.S. implementation of G-SIB surcharge framework have become regulatory priorities.
4. Emphasis on quantitative metrics has influenced regulatory reviews of supervised firms' information systems, data quality/capacity, and data management controls.
5. Emerging qualitative requirements by global supervisors focus on intraday liquidity risk management practices.
6. The emphasis of qualitative requirements has been on home country currency flows and stress testing of intraday flows.
7. Regulators are likely to expand expectations to encompass global exposures and flows for larger systemically important firms in the near term.
