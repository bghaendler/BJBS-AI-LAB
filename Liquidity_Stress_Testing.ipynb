## MEASURING CONTINGENT LIQUIDITY REQUIREMENTS

- Liquidity stress test measures the amount of liquidity an institution needs to maintain to meet financial obligations under stressed conditions.
    - "Liquidity" in this context refers to funding liquidity risk, i.e., the risk that the institution won't have enough capacity to fund its obligations without incurring economic losses.
    - Asset liquidity risk, i.e., the risk of incurring losses due to difficulty converting assets into cash, must also be considered in liquidity stress testing.
- Businesses use liquidity for four purposes:
    - operational,
    - restricted,
    - contingent, and
    - strategic.
- **Operational liquidity** is needed to fund the business on a daily basis and ensure orderly clearing of payment transactions. It is not available to meet financial obligations under a liquidity stress test.
- **Restricted liquidity** represents liquid assets available for specific purposes, and is not available to meet general financial obligations under a liquidity stress test.
- **Contingent liquidity** is available to meet general financial obligations under a stress scenario and is measured in the liquidity stress test.
- **Strategic liquidity** is held by the institution to meet future business needs and is not primarily aimed at supporting the bank during times of stress. It may be redirected to meet contingent liquidity requirements.
- Strategic liquidity is likely only feasible if it is held in highly liquid assets.

### Organizational Scope

- The consolidated stress test is essential for a liquidity risk framework.
- Stress testing can be conducted on subsidiary entities within an organization.
- A bank can stress liquidity at different organizational levels: parent, subsidiary legal entities, lines of business, service business units, and shared service centers.
- Separate liquidity stress tests may be required for each of these cases.
- For less material or manageable entities, entity-level liquidity risk reporting may be sufficient.
- The institution should consider the organizational level where liquidity is commingled and liquidity oversight has management accountability.
- Combinations of legal entities and operating units with these characteristics form the building blocks of the enterprise-level liquidity stress test.
- **Liquidity transfer restrictions.**
    - **Liquidity can be trapped in certain legal entities, leading to a distorted view of the consolidated liquidity position of the institution.**
        - Foreign exchange controls can prevent the conversion of foreign currency held in off-shore legal entities, thus hindering liquidity.
        - **The bank should assess the impact of liquidity restrictions on enterprise-level liquidity, considering normal and stressed conditions.**
        - Bank holding companies should not assume access to banking subsidiary cash during a crisis due to capital adequacy requirements.
        - The existence of liquidity transfer restrictions does not necessarily require an additional stress test, especially when subsidiaries are not required to upstream cash to the parent.
    - Institutions **may choose to stress the consolidated entity and the holding company without testing individual banking subsidiaries if movement of cash from parent to subsidiary is unrestricted**.
- **Currency**.
    - Liquidity stress tests should be performed in the currency of the entity being tested.
    - Currency conversion requirements can have a significant impact on liquidity.
    - Offshore subsidiaries or branches may carry a significant currency mismatch.
    - Settlement time frames for a home country parent to swap funds may be problematic in a crisis.
- **Regulatory jurisdiction.**
    - Institutions operating in multiple foreign jurisdictions face regulatory jurisdiction issues.
    - These institutions may need to conduct individual stress tests for foreign subsidiaries or groups.
    - U.S. regulations require certain foreign banking organizations to conduct liquidity stress tests for intermediate holding companies and branches.
    - The purpose of these stress tests is to address concerns that foreign banks operating in the country would be over-reliant on offshore funding.
- **Planning Horizon**
    - The objective of a liquidity stress test is to ensure that the institution can maintain adequate contingency funding through a period of prolonged stress.
    - The planning horizon of the liquidity stress test should be at least twelve months, although longer-term projections may be subject to significant forecast error.
    - The likelihood of the bank continuing operations indefinitely under stress without a recovery or resolution process taking place is unrealistic.
    - Banks may choose to forecast beyond twelve months if a survival horizon is calculated under the stress test.
    - The frequency of cash flow measurement within the overall time horizon must be determined, balancing the benefits of improved precision against the reduced forecasting accuracy beyond a certain time frame.
    - Stress models that forecast daily over a short time frame (e.g., one month) and transition to weekly or monthly cash flows for the remaining time horizon are likely to provide the best balance.
    - The need to forecast daily during the initial stage of the stress test is recommended, as the most critical period of stress for the institution may occur during those first few days.

## TESTING TECHNIQUES

- There are three general approaches to liquidity stress testing: historical statistical techniques, deterministic models, and Monte Carlo simulation.
    - **Historical statistical** techniques model a historical pro forma cash flow based on observed cash flow volatility, while deterministic models develop hypothetical liquidity stress scenarios based on expert judgment.
    - **Monte Carlo simulation** relies on simulation modeling and can be used to assess liquidity risk by stress testing specified variables over a future time frame.
    - **Stochastic techniques** that rely on historical observations of volatility are viewed less favorably after the financial crisis, and deterministic scenarios are viewed as the most effective tool for assessing liquidity risk.
- A challenge of stochastic approaches is their limited ability to accurately predict management countermeasures during a liquidity crisis event.
- The remainder of the chapter focuses on developing a deterministic liquidity stress test framework.

## SCENARIO DEVELOPMENT

- Liquidity failure is a rare event but can have significant impacts.
- There is limited data available to build accurate predictive models for minimum levels of liquidity.
- Liquidity stress tests involve developing a set of **deterministic scenarios.**
- The industry approach involves two types of scenarios - historical and forward-looking.
- Financial institutions should consider their unique risks when creating scenario frameworks.

### Historical Scenarios

- Historical scenarios are based on actual liquidity failures.
- They attempt to translate those events to the financial institution performing the stress test.
- Failures of WaMu and Northern Rock in 2008 are common reference events.
- Advantage: empirical basis.
- Disadvantage: few failures have occurred; limited data available.
- Future business conditions may cause new and unanticipated liquidity events, creating a potential blind spot for management.

## DEVELOPMENT OF ASSUMPTIONS

- Liquidity stress testing relies on hypothetical and historical scenarios and is highly dependent on the validity of assumptions for generating meaningful results.
- Limited historical or market data is available for many key assumptions, but applying segmentation frameworks can enhance the rigor of liquidity stress testing.
- Institutions should reference the Basel III liquidity coverage ratio cash flow rates and develop internal views of stressed liquidity behavior.
- The institution should qualitatively assess expected liquidity behavior for each type of cash flow, determine appropriate segmentation levels, assess and rank levels of liquidity risk, develop quantitative modeling assumptions based on historical data, and develop matrices of relative modeling assumptions based on scored risk levels and baseline historical data.
- Assumption matrices should be adjusted appropriately for each stress scenario.

    - The starting point for developing liquidity haircuts is a review of current market conditions and comparing them to what the bank experienced during the financial crisis. If the bank does not have such data available, it can be obtained through peer comparisons where possible.

## LIQUIDITY OPTIMIZATION

- Liquidity stress tests aim to determine the suitable size of the liquidity buffer.
- The composition of the buffer should be developed with reference to the liquidity stress test, maximizing the liquidity portfolio's efficiency.
- The Treasury can select from a range of portfolio alternatives with varying yield and capital requirements.
- The objective is to optimize the portfolio to achieve the target level of contingent liquidity required to support risk limits and regulatory requirements.
- **Liquidity versus yield:**
    - Higher yielding instruments tend to have less favorable liquidity characteristics and/or add duration to the investment portfolio.
    - Maximizing yield and/or duration may be suboptimal for an institution's return on asset performance as a whole.
    - Yield maximizing portfolios can be inefficient due to the additional balance level required to offset stress test haircuts and the mismatch between the portfolio's inflows and stress test outflows.
    - Maximizing the size of the Level 2A portfolio may reduce overall return on assets relative to investing slightly lower yielding treasury securities that require no haircut.
    - Maximizing liquidity at the expense of yield may also be inefficient.
    - It's important to have a robust liquidity stress testing model to avoid constructing a needlessly conservative liquidity portfolio.
    - Institutions should stretch the duration and risk profile of investments if the pattern of stressed net outflows demonstrates that it's comfortable to do so.
- **Liquidity vs. capital.**
    - There is a trade-off between liquidity and capital in financial institutions.
    - Incorporating the capital impact of portfolio alternatives depends on the institution's economic and regulatory capital framework.
    - Maximizing investment allocation of certain instruments, such as Level 2A agency mortgage-backed securities, may be suboptimal due to regulatory and potentially economic capital requirements.
    - Additional asset amounts may give rise to an additional equity capital requirement under leverage ratio limits.
    - Higher-risk instruments require an additional haircut, which is problematic for banks whose capital position is already strained.

## FUNDING OPTIMIZATION

- Liquidity stress testing model provides insights into the impact of different funding sources with varying liquidity characteristics.
- Since the financial crisis, financial institutions are encouraged to prefer "sticky" funding sources such as retail branch deposits over "hot" money from wholesale channels.
- Treasury should develop a target funding profile that balances liquidity and cost by explicitly modeling the liquidity impact of funding alternatives.
- Commercial deposits linked to treasury management services have a superior liquidity profile, which should encourage investment in industry segments with more intensive working capital requirements.
- An FTP framework is needed to accurately incorporate the stressed liquidity profile of various business segments across the enterprise.

## ESTABLISHING A SUSTAINABLE INFRASTRUCTURE

- A strong data management infrastructure is essential for effective liquidity stress testing.
- The infrastructure should support automated data collection, aggregation, market data capturing, report generation, and analytics.
- Many large financial institutions face challenges in developing an infrastructure for liquidity stress data due to the need for significant modifications to existing data warehouse capabilities.
- Critical requirements for a liquidity stress testing infrastructure include:
    - Integration with existing data warehouse capabilities.
    - Robust data quality and governance processes.
    - Scalability to handle large volumes of data.
    - Flexibility to adapt to changing business needs.
    - Automation of key processes to ensure efficiency and accuracy.
- **Position data collection and aggregation.**
    - A data management model should be established to collect and aggregate required liquidity position data in an automated way.
    - Different institutions can use different architectures to achieve this, such as standardized templates or data hub structures.
    - Required liquidity attributes must be associated with each position to enable automated development of model inputs.
    - Position data must conform to the required organizational granularity, which can be across multiple legal entity, business line, and jurisdictional dimensions.
    - Historic data should also be retained in the data hub for historical analysis-based model calibration.
    - The objective is to ensure that liquidity data is housed in a single location and forms a single source of truth.
- **Regulatory report generation.**
    - The goal is to generate regulatory reports with minimal manual intervention.
    - Automated position capture is a prerequisite for regulatory report generation.
    - Solutions may include functional replicas of official regulatory reports or customized proprietary versions.
    - Reports should be populated with associated validations and regulatory adjustments.
    - Minimal user manipulation should be required for these reports.
    - Solutions can be developed as proprietary applications or included within the reporting functionality of third party treasury platforms.
- **Analytics.**
    - A liquidity stress test model should have robust features and functions that meet management's expectations of a reliable analysis tool.
    - The model should be able to perform sensitivity analysis on stress test assumptions and save scenarios.
    - The model should be able to generate various legal entity views easily.
    - The leading practice is to include analytic functionality that evaluates the economic capital impact of various liquidity portfolio allocations.
- **Liquidity dashboard.**
    - A liquidity dashboard is a tool that tracks key risk indicators and performance drivers on a predefined basis.
    - The dashboard should be distributed to risk managers to ensure that they are aware of any potential risks.
    - Liquidity stress test results can be included in the dashboard or circulated separately.
    - The purpose of the liquidity dashboard is to provide a real-time overview of the liquidity position of the organization and identify any potential liquidity risks.

## INTEGRATION OF LIQUIDITY STRESS TESTING WITH RELATED RISK MODELS

- Liquidity stress testing should not be performed in isolation without considering other risk frameworks.
- Related risk frameworks such as asset liability management, capital stress testing, and recovery and resolution planning employ related assumptions concerning balance sheet behavior of certain accounts.
- Developing assumptions independently can lead to an inconsistent overall risk management framework.
- Correlations between risk types that are likely to surface in a systemic or idiosyncratic stress scenario should be considered.
- A holistic risk model that assesses the impact on liquidity, capital, and balance sheet structure under a common set of scenarios should be maintained.
- Developing unique stress scenarios for each risk type can be problematic due to modeling complexity involved.
- Interdependencies and connection points between liquidity stress model and other risk models should be carefully considered.
- **Liquidity stress testing and capital stress testing.**
    - Liquidity stress testing and capital stress testing are important for financial institutions to ensure they can withstand adverse market conditions.
    - Linking capital and liquidity stress testing requires incorporating required capital infusions of subsidiary entities into liquidity stress test scenarios.
    - Capital impact assumptions must be developed for each liquidity stress test scenario based on overall market and idiosyncratic conditions assumed to occur.
    - The level of detail in developing capital impact assumptions may vary from high-level capital infusion assumptions to detailed credit loss and pre-provision net revenue modeling.
    - The capital stress testing framework should include a liquidity stress evaluation to assess the impact of any required liquidity impacts on capital adequacy.
    - In conjunction with performing the capital stress test, the institution should perform a liquidity impact analysis to determine whether additional capital impacts may occur through investment portfolio and required funding actions that would cause further deterioration in capital adequacy.
- **Liquidity stress testing and asset liability management.**
    - Interest rate risk models are used to assess the economic value of equity impacts of severe movements in interest rates, which could have a significant impact on capital but are less likely to have a direct impact on the bank's short-term liquidity profile.
    - Liquidity stress testing models should apply a consistent behavioral framework with interest rate stress testing models.
    - The liquidity stress test scenario framework, liquidity risk dashboards, and liquidity risk early warning indicators should consider the possibility of an interest rate shock and its potential impact on indeterminate liabilities.
    - In an environment where rates are historically low and there is significant risk of a yield curve steepening, **the institution must carefully consider the impact of deposit disintermediation due to interest rate increases**, which could present significant liquidity risk to the institution.
- **Liquidity stress testing and funds transfer pricing.**
    
    
    - **Funds transfer pricing (FTP) is not a risk model, but a tool for making strategic business decisions.**
        - The proper **pricing of liquidity** is one of the key objectives of an FTP framework.
        - Liquidity provided by the treasury center for lending purposes or credited to liability-generating activities **should be priced appropriately**.
    - The FTP framework should be **consistent** with the **contingent liquidity requirement** for assets and liabilities measured by the liquidity stress test model.
    - The **cost of carrying a cash buffer** required to support a wholesale operational deposit should be passed through within the FTP framework.
    

### CONCLUSION

- Liquidity stress testing is a **crucial part of the bank's liquidity risk framework**.
- Regulators and stakeholders have **increased scrutiny and expectations for liquidity stress testing since the financial crisis**.
- Financial institutions of significant size have a basic liquidity stress testing process in place but there are **areas for improvement**.
- Banks should focus on four areas to refine and improve their liquidity stress capabilities:
    1. Ensuring the **appropriate scope and structure of the liquidity stress test**.
    2. Building the **model on robust assumptions**.
    3. Improving **integration with related risk and performance models**.
    4. **Automating the process** to create a sustainable liquidity risk infrastructure.
- The **industry is moving towards unifying liquidity risk, capital, asset liability management, and performance measurement frameworks**.
